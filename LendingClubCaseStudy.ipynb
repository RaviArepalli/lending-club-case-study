```python
## Identifying the variables 
### We know that the dataset have only past values of 3 kinds Full paid, Charged off, and default So we can now safely
### remove the charged off / current loan status 
```

```python
df = df[~(df["loan_status"]=='Current')]
df.head()
```

```python
# Removing Null columns 
df.isnull().sum()
```

```python
round(df.isnull().sum()/len(df.index), 3)*100
```

```python
#Names of null columns that are getting removed 
all_null_columns = [x for x in df if df[x].isna().sum() == df.shape[0]]
print("Null columns: ", all_null_columns)
```

```python
df.drop(all_null_columns, inplace=True, axis=1)
```

```python
round(df.isnull().sum()/len(df.index), 2)*100
```

```python
### Further clean up of data by removing null columns 
df.drop(["desc", "mths_since_last_delinq", "mths_since_last_record"], inplace=True, axis=1)
```

```python
round(df.isnull().sum()/len(df.index), 3)*100
```

```python
### Now, we are going to find the unique columns to help us sort categorical variables
df[df.columns].nunique()
```

```python
## After seeing the data we came to conclusion that we can use id as the unique identifers for the data rows
### So we are dropping columns that can be used but not requried as of now to derive the conclusion 
df.drop(["url", "member_id"], inplace=True, axis=1)
```

```python
### Removing behaviour variables that can help new users also to apply for the loan 
df.drop([
    'delinq_2yrs', 
    'earliest_cr_line', 
    'inq_last_6mths', 
    'open_acc', 
    'pub_rec', 
    'revol_bal', 
    'revol_util',
    'total_acc',
    'out_prncp',
    'out_prncp_inv',
    'total_pymnt',
    'total_pymnt_inv',
    'total_rec_int',
    'total_rec_prncp',
    'total_rec_late_fee',
    'recoveries',
    'collection_recovery_fee',
    'last_pymnt_d',
    'last_pymnt_amnt',
    'last_credit_pull_d'], axis=1, inplace=True)
```

```python
df.nunique().sort_values().head(15)
```

```python
df.drop(['pymnt_plan', "initial_list_status",'collections_12_mths_ex_med','policy_code','acc_now_delinq', 'application_type', 'pub_rec_bankruptcies', 'tax_liens', 'delinq_amnt', 'chargeoff_within_12_mths'], axis = 1, inplace = True)
df.head()
```

```python
### Now lets see the data info
df.info()
```

```python
### We need to clean the int_rate as it is showing object so it seems there is some correction required in this 
### before proceeding 
### Cleaning of int_rate
df['int_rate'].head(10)
```

```python
### Cleaning of int_rate
df['int_rate'] = df['int_rate'].str.rstrip('%').astype('float')
```

```python
df.info()
```

```python
df.home_ownership.value_counts()
```

```python
##Replacing None with other so that we can reduce the variables
df['home_ownership'].replace(to_replace = ['NONE'],value='OTHER',inplace = True)
df.home_ownership.unique()
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'],format='%b-%y')
```

```python
df['MonthIssued'] = df.issue_d.dt.month
df['YearIssued'] = df.issue_d.dt.year
```
